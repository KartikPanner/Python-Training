#### Handle the missing values in the column Percent_Quantity_With_Discount with strategy = mean 

```python
from sklearn.impute import SimpleImputer
```

```python
imp =SimpleImputer(missing_values=np.nan, strategy ='mean')
imp.fit(df1['Percent_Quantity_With_Discount'].values.reshape(-1,1))
df1['Percent_Quantity_With_Discount']=pd.DataFrame(imp.transform(df1['Percent_Quantity_With_Discount'].values.reshape(-1,1)))
```
