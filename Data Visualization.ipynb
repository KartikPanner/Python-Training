### Heatmap : Heatmap is a graphical representation of data that uses color-codings to represent different values. It take two dimensional dataset as parameter

#### As the values in the Airfares dataframe are contrast I have just taken two columns as example to depict with the Heatmap

#### The values in heatmap are actually mapped to specific color. In the below chart, darkest color is representing the lowerst value and lightest color represents the highest value

```python
Tab_af2=Airfares[['Departure_avg_INCOME','Arrival_avg_INCOME']]
Tab_af3=Airfares[['Departure_avg_population','Arrival_avg_population']]
```

#### Below I have used the flight_data that was used in the previous session to better depict the heatap.
#### In this example, we can see clear usage of colors for the differet values, in which dark color represents the high value and light color represents the lower value.


#### Also we have used ' annot '  parameter which prints the value of each box with its corresponding values, by just giving annot = True the heatmpa will be printed with the values. This helps us to undestand better on the graph and its values. We can change the font size, weight, text etc., using annot_kws ( keywords).

#### Another parameter that was used is 'cmap', this is nothing but a parameter that allows us to customize our colors or color pallets in heatmap. Usig ths parameter, we can choose to use a single color or multi color to represent the data. Below is a representation of multi-color pallet, where blue represents positive value and brown represents negative values in the color map.

### Regplot : Regplot stands for Regression plot and the purpose is to build and visualize linear regression of our data.

#### Below is the example of comaprison of the data Distance & Fare with the correlation of linear regression. i.e., regplot helps to understand the linear relationship between x & y variables

#### The shaded bands in the around the line are actually confidence intervals. Seaborn uses a process called boostrapping to create this confidence intervals

```python
sns.regplot(x=Airfares['DISTANCE'],y=Airfares['FARE'])
```

### lmplot : lmplot stands for Linear Modal plot
#### Throught this plot modal, we can come to know how this linear line relates between x and y. As like regplot, this line is created by the same process called bootstrapping

```python
sns.lmplot(x='DISTANCE',y='FARE',data=Airfares)
```

#### When we use hue parameter in this chart, it actually provides the linear modal of the variables passsed and that's the reason we are seeing two linear regression lines below.

```python
sns.lmplot(x='DISTANCE',y='FARE',data=Airfares,hue='SLOT')
```

#### We can also include a parameter ' COL ' to get the above out-put for better understanding on the variables passed 

```python
sns.lmplot(x='DISTANCE',y='FARE',data=Airfares,hue='SLOT',col='SLOT')
```

### Pairplot : The pairplot function creates a grid of Axes such that each variable in data will by shared in the y-axis across a single row and in the x-axis across a single column. It generates multiple plots with the data and show it in a consolidated format.

#### In the below example, we can notice that all the variables were actually compared with itself and other variables of the dataframe and given us the output in a consolidated format

```python
sns.pairplot(data=Airfares)
```
